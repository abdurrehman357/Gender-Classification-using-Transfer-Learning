```python
import matplotlib.pyplot as plt
```

```python
# Load the results of each model
model1_results = [0.85, 0.89, 0.92]
model2_results = [0.82, 0.87, 0.91]
model3_results = [0.80, 0.85, 0.89]
```

```python
# Print the data arrays
print(model1_results)
print(model2_results)
print(model3_results)
```

```python
# Set the x-axis labels
x_labels = ['Epoch 1', 'Epoch 2', 'Epoch 3']
```

```python
# Set the plot size
plt.figure(figsize=(8, 6))
```

```python
# Plot the results
plt.plot(x_labels, model1_results, label='Model 1')
plt.plot(x_labels, model2_results, label='Model 2')
plt.plot(x_labels, model3_results, label='Model 3')
```

```python
# Set the plot title and axis labels
plt.title('Accuracy Results of Three Models')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

```python
# Add a legend to the plot
plt.legend()
```

```python
# Show the plot
plt.show()
```
